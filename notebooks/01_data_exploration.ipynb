# 01 — Data Exploration

This notebook explores the raw intraday payment transactions dataset.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.data_processing import load_raw_data

%matplotlib inline
sns.set_theme(style='whitegrid')
```

## 1. Load raw data

```python
df = load_raw_data('../data/raw/synthetic_intraday_payments_500k.csv')
print(f'Shape: {df.shape}')
df.head()
```

## 2. Basic statistics

```python
df.describe(include='all').T
```

## 3. Transaction type distribution

```python
fig, ax = plt.subplots(figsize=(6, 4))
df['transaction_type'].value_counts().plot(kind='bar', ax=ax, color=['steelblue', 'salmon'])
ax.set_title('Transaction Type Distribution')
ax.set_xlabel('Transaction Type')
ax.set_ylabel('Count')
plt.tight_layout()
plt.savefig('../figures/transaction_type_dist.png', dpi=150)
plt.show()
```

## 4. Hourly volume distribution

```python
df['hour'] = pd.to_datetime(df['timestamp']).dt.hour
fig, ax = plt.subplots(figsize=(10, 4))
df.groupby('hour')['amount'].sum().plot(kind='bar', ax=ax, color='steelblue')
ax.set_title('Total Transaction Volume by Hour of Day')
ax.set_xlabel('Hour')
ax.set_ylabel('Total Amount (USD)')
plt.tight_layout()
plt.show()
```

## 5. Amount distribution (log scale)

```python
import numpy as np
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(np.log1p(df['amount']), bins=60, color='steelblue', edgecolor='white')
ax.set_title('Log-Amount Distribution')
ax.set_xlabel('log(1 + amount)')
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## 6. Transactions per bank

```python
fig, ax = plt.subplots(figsize=(12, 4))
df['bank_id'].value_counts().plot(kind='bar', ax=ax, color='teal')
ax.set_title('Transaction Count per Bank')
ax.set_xlabel('Bank ID')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```
